### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO

import numpy as np
import random
import matplotlib.pyplot as plt
import statistics as stats
import pandas as pd

x = np.random.normal(5.0, 1.0, 100)
lenght1 = len(x)
print("Lenght of the first array:", lenght1, "\n")

x2 = np.random.chisquare(df=3, size=(100))
lenght2 = len(x2)
print("Lenght of the second array:", lenght2, "\n")

print("________________________________________________________________________________________________________________________________", "\n")

mean1 = np.mean(x)
print("The mean for the first array is:", mean1, "\n")
mode1 = stats.mode(x)
print("The mode for the first array is:", mode1, "\n")
median1 = np.median(x)
print("The median for the first array is:", median1, "\n")

mean2 = np.mean(x2)
print("The mean for the second array is:", mean2, "\n")
mode2 = stats.mode(x2)
print("The mode for the second array is:", mode2, "\n")
median2 = np.median(x2)
print("The median for the second array is:", median2, "\n")

print("________________________________________________________________________________________________________________________________", "\n")

range1 = max(x) - min(x)
print("The difference (range) of the max value against the lowest value of the first array is:", range1, "\n")
variance1 = stats.variance(x)
print("The variance of the first array is:", variance1, "\n")
std_dev1 = stats.stdev(x)
print("One standard deviation of the first array equals to:", std_dev1, "\n")

range2 = max(x2) - min(x2)
print("The difference (range) of the max value against the lowest value of the second array is:", range1, "\n")
variance2 = stats.variance(x2)
print("The variance of the second array is:", variance2, "\n")
std_dev2 = stats.stdev(x2)
print("One standard deviation of the second array equals to:", std_dev2, "\n")

print("________________________________________________________________________________________________________________________________", "\n")

percentile90_1 = np.quantile(x, 0.90)
print("The percentile 90 for the first array is:", percentile90_1, "\n")
Q1 = np.quantile(x, 0.25)
print("The first quartile of the first array is:", Q1, "\n")
Q2 = np.quantile(x, 0.5)
print("The second quartile of the first array is:", Q2, "\n")
Q3 = np.quantile(x, 0.75)
print("The third quartile of the first array is:", Q3, "\n")
IQR = Q3 - Q1
print("The interquantile range (IQR) for the first array is:", "\n", "\n")

percentile90_2 = np.quantile(x2, 0.90)
print("The percentile 90 for the second array is:", percentile90_2, "\n")
Q1_2 = np.quantile(x2, 0.25)
print("The first quartile of the second array is:", Q1_2, "\n")
Q2_2 = np.quantile(x2, 0.5)
print("The second quartile of the second array is:", Q2_2, "\n")
Q3_2 = np.quantile(x2, 0.75)
print("The third quartile of the second array is:", Q3_2, "\n")
IQR_2 = Q3_2 - Q1_2
print("The interquantile range (IQR) for the second array is:", "\n", "\n")

print("________________________________________________________________________________________________________________________________", "\n")

x_df = pd.DataFrame(x) #First array turned dataframe
x2_df = pd.DataFrame(x2) #Second array turned dataframe

asimetry1 = x_df.skew()
print("The skewness of the data of the first DataFrame is:", asimetry1, "\n")
kurtosis1 = x_df.kurtosis()
print("The kurtosis of the data of the first DataFrame is:", kurtosis1, "\n", "\n")

asimetry2 = x2_df.skew()
print("The skewness of the data of the second DataFrame is:", asimetry2, "\n")
kurtosis2 = x2_df.kurtosis()
print("The kurtosis of the data of the second DataFrame is:", kurtosis2, "\n", "\n")
````
